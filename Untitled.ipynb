Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
```

```python
# Number of observations in each class 
N1 = np.random.randint(1,10.0001)*100 
N2 = np.random.randint(1,10.0001)*100 
N3 = np.random.randint(1,10.0001)*100
```

```python
#excel table representation
df = pd.read_excel('table.xlsx')
df.head()
```

Claster 1 the Ranges from 800>>>12000 for #Steps per Day and 7>>>9 for hours Slept the same anology applies for Cluster 2 and 3 
So this information will be used to create observations for each and every cluster.

```python
Cluster1_steps = np.random.randint(8000, 12001, N1)
Cluster2_steps = np.random.randint(5000, 8001, N2)
Cluster3_steps = np.random.randint(2000, 5001, N3)
```

Above is for the steps

```python
Cluster1_sleep = np.random.uniform(7, 9.00001, N1)
Cluster2_sleep = np.random.uniform(6, 7.50001, N2)
Cluster3_sleep = np.random.uniform(5, 6.50001, N3)
```

```python
df_cluster1 = pd.DataFrame({'Steps_per_Day': Cluster1_steps, 'Hours_Slept_per_Night': Cluster1_sleep, })
df_cluster2 = pd.DataFrame({'Steps_per_Day': Cluster2_steps, 'Hours_Slept_per_Night': Cluster2_sleep, })
df_cluster3 = pd.DataFrame({'Steps_per_Day': Cluster3_steps, 'Hours_Slept_per_Night': Cluster3_sleep, })

# Combine all clusters
combine = pd.concat([df_cluster1, df_cluster2, df_cluster3], ignore_index=True)
```

A plot 

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_cluster1['Steps_per_Day'], df_cluster1['Hours_Slept_per_Night'], color='blue',label="Active", alpha=0.5, edgecolors='black')
plt.scatter(df_cluster2['Steps_per_Day'], df_cluster2['Hours_Slept_per_Night'], color='green',label="Moderately active", alpha =0.5, edgecolors='black')
plt.scatter(df_cluster3['Steps_per_Day'], df_cluster3['Hours_Slept_per_Night'], color='red',label="Least active", alpha =0.5, edgecolors='black')


plt.xlabel("Steps per Day")
plt.ylabel("Hours Slept per Night")
plt.title("Fitness Tracker Data Visualization")

plt.legend(title="Clusters")

# Show Plot
plt.show()
```

K-means Clustering Representa

```python
# Define the number of clusters (should be 3 for this dataset)
n_clusters = 3 
# Create a KMeans object with the specified number of clusters 
kmeans = KMeans(n_clusters=n_clusters,random_state=42) 
# Fit the KMeans model to the data
kmeans.fit(combine) 
# Get the cluster labels assigned by KMeans 
kmeans_labels = kmeans.labels_ 
print("Cluster labels for each data point:\n", kmeans_labels)
```

```python
colors = ['blue', 'green', 'red']  # Choose different colors for each cluster
```

```python
plt.figure(figsize=(10, 6))

for cluster in range(3):  # Loop through each cluster
    plt.scatter(
        combine['Steps_per_Day'][kmeans_labels == cluster],  # Select x values for current cluster
        combine['Hours_Slept_per_Night'][kmeans_labels == cluster],  # Select y values for current cluster
        color=colors[cluster],  # Assign cluster-specific color
        label=f"Cluster {cluster + 1}",  # Label for the legend
        edgecolor='black',  # Black edges for visibility
        alpha=0.5  # Transparency for better visualization
    )

# Add labels and title
plt.xlabel("Steps per Day")
plt.ylabel("Hours Slept per Night")
plt.title("KMeans Clustering of Fitness Tracker Data")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Create a DataFrame with the cluster labels
combine['Cluster'] = kmeans_labels

# Compute average values for each cluster
cluster_analysis = combine.groupby('Cluster').agg(
    avg_steps=('Steps_per_Day', 'mean'),
    avg_sleep=('Hours_Slept_per_Night', 'mean'),
    count=('Steps_per_Day', 'count')  # Number of points in each cluster
)

# Print the analysis
print("Cluster Analysis (Average Steps & Sleep per Cluster):")
print(cluster_analysis)
```

This code analyzes and visualizes clusters formed by the KMeans algorithm, focusing on Steps per Day and Hours Slept per Night.

```python
# Get cluster indices
clusters = cluster_analysis.index
avg_steps = cluster_analysis['avg_steps']
avg_sleep = cluster_analysis['avg_sleep']

# Create the figure and axes for subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot for Average Steps per Cluster
ax[0].bar(clusters, avg_steps, color=['blue', 'green', 'red'])
ax[0].set_title("Average Steps per Cluster")
ax[0].set_xlabel("Cluster")
ax[0].set_ylabel("Avg Steps per Day")
ax[0].set_xticks(clusters)

# Bar plot for Average Sleep per Cluster
ax[1].bar(clusters, avg_sleep, color=['blue', 'green', 'red'])
ax[1].set_title("Average Sleep per Cluster")
ax[1].set_xlabel("Cluster")
ax[1].set_ylabel("Avg Hours Slept per Night")
ax[1].set_xticks(clusters)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

###Three distinct clusters were identified:

Active group (Blue Cluster): High steps (9000+ per day) and adequate sleep (7–9 hours). Represents health-conscious individuals.

Moderately active group (Green Cluster): Moderate steps (approx 6000 per day) and sleep (approx 6–7.5 hours). Likely busy professionals or casual exercisers.

Least active group (Red Cluster): Low steps (<5000 per day) and insufficient sleep (<6.5 hours). Indicates a sedentary lifestyle, potentially unhealthy habits.

###*Key Findings:*

Higher activity levels are associated with better sleep.

The least active group may require interventions for better health habits.

The data can be used for personalized fitness & sleep improvement recommendations.
